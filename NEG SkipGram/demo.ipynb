```python
from word2vec.utils import load_embeddings, plot_embeddings, get_list_of_interesting_words


words, embeddings = load_embeddings('data/word_embeddings_30.txt')
```

```python
# load the list of interesting words that we want to plot
interesting_words = get_list_of_interesting_words()

interesting_indices = [i for i, word in enumerate(words) if word in interesting_words]

selected_words = [words[i] for i in interesting_indices]
selected_embeddings = embeddings[interesting_indices]
```

```python
plot_embeddings(selected_embeddings, selected_words)
```
